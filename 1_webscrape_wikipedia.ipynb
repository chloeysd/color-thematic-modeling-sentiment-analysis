```python
import requests
from bs4 import BeautifulSoup

def extract_main_body_text(page_title, language='en'):
    #add a user agent to simulate real user in order to avoid being blocked due to too many requests, as Ref:https://www.scrapehero.com/how-to-prevent-getting-blacklisted-while-scraping/
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.61 Safari/537.36'}
    wikipedia_url = f'https://{language}.wikipedia.org/wiki/{page_title}'
    
    response = requests.get(wikipedia_url, headers=headers)
    if response.status_code != 200:
        return None
    
    soup = BeautifulSoup(response.text, 'html.parser')

    #Empty string to put our content
    main_body_text = ""
    #Use BeautifulSoup's find_all method to find all <p> tags (paragraphs) on the webpage
    paragraphs = soup.find_all('p')

    # If we have retrieve content
    if paragraphs:
    # Find each paragraph
      for paragraph in paragraphs:
          # And add that paragraph to our main_body_text string
          main_body_text += paragraph.get_text()
          # Add a new line after each paragraph
          main_body_text += '\n'

    return main_body_text   

#scrape data for ten colors from five domains.
colors_topics = {
    "red": [#colour：
         "Red","CMYK color model","RGB color model","Scarlet (color)","Vermilion",
          #Art & culture:
         "Art history","Prehistoric art","Renaissance art","Flag","China", "Culture of ancient Rome","Mayan civilization",
          #Science and nature:
          "Mars","Great Red Spot","Red dwarf","Visible spectrum","Sunrise","Laser",
          #Psychology and symbolism:
          "Color psychology","Red culture movement","Emotion","Heat","Anger","War",
          #sport：
          "chariot racing","association football","Scuderia Ferrari","Sport psychology","Color commentator"
          ] ,
    "blue": [
         #colour：
         "Blue","Primary color","RGB color model","Violet (color)","Azure (color)","Ultramarine","Sky blue",
          #Art & culture:
         "Art history","Middle Ages","Blue and white porcelain","Flag","Santorini","Shah Mosque (Isfahan)",
          #Science and nature:
          "Blue LED", "Atmospheric perspective","Algae","Rayleigh scattering","Halo (optical phenomenon)","Tyndall effect",
          #Psychology and symbolism:
          "Color psychology","Sky","Sea","Logan Sapphire","sad","Uniform",
          #sport：
          "Oceania Football Confederation","basketball","Argentina"
          ],
    "green":  [#colour：
          "Green","Subtractive color","RGB color model","Primary color",
          #Art & culture:
          "Art history","Post-classical history","Modern art","Flag","Mesopotamia","hunter green","Egyptian hieroglyphs",
          #Science and nature:
          "plant","photosynthesize","Nature","chlorophyll","Emerald",
          #Psychology and symbolism:
          "Color psychology","Health","Emotion","Youth","Hope","Envy","Life" ,"traffic lights",         
          #sport：
          "Judo","Casino","British racing green"
          ],
    "orange": [#colour：
          "Orange (colour)","Secondary color","RGB color model",
          #Art & culture:
          "Art history","Color theory","Flag"," Ancient India","Ancient Egypt","National colours",
          #Science and nature:
          "Spice","Orpiment","Autumn leaf color","Carrot","Pumpkin","Arches National Park",
          #Psychology and symbolism:
          "Color psychology","Safety orange","Prisoners","flight data recorder","Halloween",
          #sport：
          "Bandy balls","Netherlands national football team","Boardslides"
          ]  ,
    "yellow" :[#colour：
          "Yellow","Subtractive color","Ochre","Primary color","RGB color model",
          #Art & culture:
          "Art history","Byzantine art","Post-classical history","Flag","China","Egypt","Las Vegas",
          #Science and nature:
          "Egg yolks", "Buttercups", "Corn","Sunlight","Acacia dealbata","Indian yellow",
          #Psychology and symbolism:
          "Color psychology","Light","Gold","Safety yellow","optimism","pleasure","Taxicabs",
          #sport：
          "Penalty card","Yellow jersey","Tour de France"
          ],
    "purple":[#colour：
          "Purple","Violet (color)","Additive color","Spectral color",
          #Art & culture:
          "Art history","Impressionism","Tyrian purple","Flag","China","Byzantine Empire",
          #Science and nature:
          "Silk","Blackberry","Manganese violet", "Hematite","Dye","Purple finch","Purple heron",
          #Psychology and symbolism:
          "Color psychology","Eroticism","Femininity","Seduction","Polaris",
          #sport：
          "National Basketball Association","Indian Premier League","Baltimore Ravens","Fremantle Football Club"
          ] ,
    "black" : [#colour：
          "Black","Hue","Grey",
          #Art & culture:
          "Art history","Dark Ages (historiography)","Neolithic","Roman Empire","Anubis Shrine","Age of Enlightenment","Middle Ages",
          #Science and nature:
          "Night", "Underworld","Carbon nanotube","Light","Pigment","Ultraviolet","Blacklight","Infrared",
          #Psychology and symbolism:
          "Color psychology","Evil","Death","Witchcraft","Magic (supernatural)","Mourning","Darkness","Black Friday (shopping)",
          #sport：
          "Rugby union","Blackballing"
          ] ,
    "cyan" : [#colour：
          "Cyan","Visible spectrum","CMYK color model","Turquoise (color)","Electric blue (color)","Aquamarine (color)","Teal",
          #Art & culture:
          "Art history","Ancient Greek","Cyanotype","Cinecolor",
          #Science and nature:
          "Centaurea cyanus","water","Natural gas","Uranus","Flame","Lapis lazuli","Prussian blue",
          #Psychology and symbolism:
          "Color psychology","Cooking","Gas stove","Bacteria","Air",
          ]  ,
    "pink":[#colour：
          "Pink","Rose (color)","Shades of magenta",
        #Art & culture:
         "Art history","Madonna of the Pinks","Emma, Lady Hamilton","Lipstick","Barbie",
          #Science and nature::
          "Dianthus plumarius", "Rose","Flamingo","Rhodochrosite","Flowering plant","Blossom",
          #Psychology and symbolism:
          "Color psychology","Gendered associations of pink and blue","Gender","Toy",
          #sport：
          "Palermo FC","Cerezo Osaka","Inter Miami CF",
          ]   ,
     "white": [#colour：
         "White","Hue","Visible spectrum","Primary color",
         #Art & culture:
         "Art history","Roman temple","Ancient Greek temple","Neoclassical architecture","Modern architecture",
          #Science and nature::
          "Electromagnetic spectrum", "White point","Cumulus cloud","Hyams Beach, New South Wales","Beluga whale","Penguin",
          #Psychology and symbolism:
          "Color psychology","Cleanliness","Ghost","Pearl","Snow","Milk","Chalk",
          ]   
}

#a nested loop to collect data for each color and topic.
for color, topics in colors_topics.items():
    for topic in topics:
        #the function accepts topics and 'en' as parameters, where spaces are replaced with underscores.
        main_body_text = extract_main_body_text(topic.replace(" ", "_"), language='en')
        if main_body_text:
            file_path = f"//Users/yinshuodi/Desktop/mini-project-23006440/Webcrawling/code&data/data/my-data/{color}/{topic}-wikipedia-text.txt"
            with open(file_path, "w") as myfile:
               myfile.write(main_body_text)
```
